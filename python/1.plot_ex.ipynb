```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.normal(loc=4.1, scale=2.2, size=1000)
```

```python
type(x)
```

```python
h_fig = plt.hist(x, bins=25)
plt.xlabel('Simulated sample')
plt.ylabel('frequency')
plt.show()
```

```python
xts = pd.Series(x)
type(xts)
```

```python
xts.plot()
plt.xlabel('Time')
plt.ylabel('Simulated sample')
plt.show()
```
